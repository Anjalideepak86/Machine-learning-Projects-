# PIE CHART

```python
plt.figure(figsize=(10,10))
colors = ['red','yellow','green','blue',"orange","magenta"]

# Aggregate sales by customers 
customer_sales =df.groupby("LIFESTAGE")["TOT_SALES"].sum()

# Create pie chart
plt.pie(customer_sales, labels=customer_sales.index, autopct="%1.1f%%", colors=colors,shadow=True)
plt.title("Revenue Contribution by Customers")

# Show the plot
plt.axis("equal")
plt.show()

```

```python
customer_sales = df.groupby(["LIFESTAGE", "PREMIUM_CUSTOMER"])["TOT_SALES"].sum().reset_index()
customer_sales
```

```python
top_customers =customer_sales.sort_values(by="TOT_SALES", ascending=False)
top_customers
```

```python
top_customers.head(10)
```

```python
#Total sales by profitable  product
top_products =df.groupby('PROD_NAME')['TOT_SALES'].sum().reset_index()
top_products
```

```python
top_products_sorted=top_products.sort_values(by='TOT_SALES',ascending=False)
top_products_sorted
```

```python
top_products.head(3)
```

```python
loyal_customers =df.groupby(['LIFESTAGE','PREMIUM_CUSTOMER']).agg(total_spend=("TOT_SALES","sum"),unique_numbers=("LYLTY_CARD_NBR","nunique"),avg_spend_by_customers=("TOT_SALES","mean")).sort_values(by="total_spend",ascending=False)
print(loyal_customers.head(3))
```
